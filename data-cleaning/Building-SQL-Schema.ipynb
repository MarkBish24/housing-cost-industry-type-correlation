```python
import pandas as pd
import glob
import os

# Get all files

all_files = glob.glob(os.path.join('./Industry-data-clean', "*.csv"))

df_list = []

year_cols = [str(year) for year in range(2010, 2026)]

for file in all_files:
    county_name = os.path.splitext(os.path.basename(file))[0]

    df = pd.read_csv(file)

    df_long = df.melt(id_vars=['TITLE'], value_vars=year_cols,
                      var_name='Year', value_name='Value')

    df_long['County'] = county_name

    df_long['Year'] = df_long['Year'].astype(int)

    df_list.append(df_long)
    
# Combine all counties into one long-format DataFrame
df_all = pd.concat(df_list, ignore_index=True)

# Reorder columns
df_all = df_all[['County', 'TITLE', 'Year', 'Value']]

df_all.head()
```

```python
df_all.tail()
```

```python
df_all['County'] = df_all['County'].str.replace('_per_mil', '', regex=False)
df_all['County'] = df_all['County'].str.replace('_', ' ', regex=False)

# Clean IndustryType names
df_all['TITLE'] = df_all['TITLE'].str.replace(' per mil', '', regex=False)

df_all.head(30)
```

```python
df_all = df_all.rename(columns={
    'TITLE': 'Industry',
    'Value': 'PerMilValue'
})
df_all
```

```python
df_counties = pd.read_csv("./Schema-data/Counties.csv")  # Columns: ID, RegionName
df_counties = df_counties.rename(columns={'id': 'county_id', 'RegionName': 'County'})

df_industry = pd.read_csv("./Schema-data/Industries.csv")  # Columns: ID, IndustryType
df_industry = df_industry.rename(columns={'id': 'industry_id', 'Industry': 'Industry'})
```

```python
df_all = df_all.merge(df_counties[['county_id', 'County']], on='County', how='left')

df_all = df_all.merge(df_industry[['industry_id', 'Industry']],on='Industry', how='left')
```
