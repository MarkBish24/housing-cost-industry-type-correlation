```python
df = pd.read_csv("./Housing-data-clean/combined_file.csv")
df.head()
```

```python
year_cols = [str(year) for year in range(2000, 2026)]

df_long = df.melt(id_vars=['RegionName'], value_vars=year_cols, 
                  var_name='Year', value_name='HousingCost')
# Optional: convert Year to integer
df_long['Year'] = df_long['Year'].astype(int)

df_counties= pd.read_csv("./Schema-data/Counties.csv")

df_long = df_long.merge(df_counties[['id', 'RegionName']], on='RegionName', how='left')

#df_long['HousingID'] = ['H' + str(i).zfill(4) for i in range(1, len(df_long)+1)]

df_long = df_long.rename(columns={'id':'county_id'})

df_long['id'] = ['H' + str(i).zfill(4) for i in range(1, len(df_long)+1)]

df_long = df_long[['id', 'county_id', 'Year', 'HousingCost']]

df_long.head(50)
```

```python
df_long.to_csv("./Schema-data/Housing-Cost.csv", index=False)
```
