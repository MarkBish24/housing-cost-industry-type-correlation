```python
import pandas as pd

# Just read it directly, no os.makedirs needed
df = pd.read_csv(".\Housing-data-raw\Zillow_County_SFH.csv")
df.head()
```

```python
ca_counties = df[df["State"] == "CA"]
ca_counties
```

```python
only_dates = ca_counties.iloc[:, 9:]  # slice housing data (date columns)
only_dates.columns = pd.to_datetime(only_dates.columns, errors="coerce")  # convert col names to datetime
only_dates = only_dates.groupby(only_dates.columns.year, axis=1).mean().round(0)  # group by year
only_dates
```

```python
only_info = ca_counties[["RegionName", "RegionType", "State", "Metro"]]
only_info
```

```python
combined = pd.concat([only_info, only_dates], axis=1)
combined
```

```python
combined.to_csv("./Housing-data-clean/combined_file.csv", index=False)
```
