```python
import pandas as pd
import os
import re

# Data Paths
input_folder = "./Industry-data-raw"
output_folder = "./Industry-data-clean"

# Make sure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Load the Excel file into a DataFrame
for file_name in os.listdir(input_folder):
    if file_name.endswith(".xls") or file_name.endswith(".xlsx"):  # only Excel files
        # Build full file path
        file_path = os.path.join(input_folder, file_name)
        
        # Read Excel file
        df = pd.read_excel(file_path)

        # Getting file name-County name
        county_name = df.iloc[0, 6]
        if "county" not in county_name.lower():
            county_name = df.iloc[1, 6]

        county_name = re.sub(r"[()]", "", county_name).replace(" ", "_")

        # Removing unnecesary info
        df = df.iloc[6:].reset_index(drop=True)
        df.columns = df.iloc[0]
        df = df[1:].reset_index(drop=True)

        # Seperating by dates and titles
        only_dates = df.iloc[:, 8:]
        only_info = df['TITLE']

        # Organize info by years instead of months and averaged them out
        only_dates.columns = pd.to_datetime(only_dates.columns)
        df_yearly = only_dates.groupby(only_dates.columns.year, axis=1).mean().round(0)

        # Combine them back together
        df_combined = pd.concat([only_info, df_yearly], axis=1)
        df_combined = df_combined.iloc[:-9].reset_index(drop=True)

        # remove the row where Title == "Civilian Unemployment Rate"
        df_combined = df_combined[df_combined["Title"] != "Civilian Unemployment Rate"]

        df_per_mil = df_combined.copy()
        clf = df_per_mil[df_per_mil['TITLE'].str.contains("Civilian Labor Force", case=False)].iloc[0, 1:]

        for col in df_per_mil.columns[1:]:  # skip 'TITLE'
            df_per_mil[col] = (df_per_mil[col] / clf[col]) * 1_000_000

        # Export data in clean folder
        df_combined.to_csv(f"./Industry-data-clean/{county_name}.csv", index=False)

# EVERY THING BELOW IS TESTED CODE ^^^ ABOVE IS A FOR LOOP DOING THE SAME THING ^^^
```
